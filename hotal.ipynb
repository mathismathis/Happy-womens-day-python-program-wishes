```python
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding, Attention
from tensorflow.keras.models import Model
import numpy as np



# Preprocess the data and create vocabulary
def preprocess_data(hotel_data):
    # Combine all customer reviews into a single list
    reviews = []
    for hotel in hotel_data["Hotels"]:
        for review in hotel["CustomerReviewData"]:
            reviews.append(review["ReviewText"])

    print(reviews)

    # Tokenize the reviews and create vocabulary
    tokenizer = tf.keras.preprocessing.text.Tokenizer(oov_token="<UNK>")
    tokenizer.fit_on_texts(reviews)
    vocab_size = len(tokenizer.word_index) + 1

    return tokenizer, vocab_size

# Encode text data using the tokenizer
def encode_text(tokenizer, texts, max_len):
    return tf.keras.preprocessing.sequence.pad_sequences(
        tokenizer.texts_to_sequences(texts), maxlen=max_len, padding="post"
    )

# Create seq2seq model with attention
def create_seq2seq_model(vocab_size, embedding_dim, hidden_units, max_len):
    # Encoder
    encoder_inputs = Input(shape=(max_len,))
    encoder_embedding = Embedding(vocab_size, embedding_dim)(encoder_inputs)
    encoder_outputs, state_h, state_c = LSTM(hidden_units, return_state=True)(encoder_embedding)
    encoder_states = [state_h, state_c]

    # Decoder
    decoder_inputs = Input(shape=(None,))
    decoder_embedding = Embedding(vocab_size, embedding_dim)(decoder_inputs)
    decoder_lstm = LSTM(hidden_units, return_sequences=True, return_state=True)
    decoder_outputs, _, _ = decoder_lstm(decoder_embedding, initial_state=encoder_states)

    # Attention mechanism
    attention_layer = Attention()
    attention_outputs = attention_layer([decoder_outputs, encoder_outputs])

    # Output layer
    decoder_dense = Dense(vocab_size, activation='softmax')
    decoder_outputs = decoder_dense(attention_outputs)

    model = Model([encoder_inputs, decoder_inputs], decoder_outputs)
    return model

# Load the tokenizer and vocabulary size
tokenizer, vocab_size = preprocess_data(hotel_data)

# Example usage:
embedding_dim = 256
hidden_units = 512
max_len = 100  # Replace with an appropriate value

seq2seq_model = create_seq2seq_model(vocab_size, embedding_dim, hidden_units, max_len)

# Train the model using hotel-related data (you may need more data for better performance)
# ...

# Use the trained model for inference to generate responses to user queries
def chatbot_response(query):
    query_encoded = encode_text(tokenizer, [query], max_len)
    response = seq2seq_model.predict([query_encoded, np.zeros((1, 1))])
    response_text = tokenizer.sequences_to_texts([np.argmax(response[0], axis=-1)])
    return response_text[0]

# Example usage:
user_query = "What are the room prices at Sunny Beach Resort?"
response = chatbot_response(user_query)
print(response)  # Output the chatbot response
```

```python
import pandas as pd
import tensorflow as tf
import tensorflow_hub as hub

# Load your question-answer dataset (replace 'your_data.csv' with the path to your dataset)
df = pd.read_csv('hotel.csv')

# Preprocess the questions and answers (lowercase and remove punctuation)
def preprocess_text(text):
    if isinstance(text, str):  # Check if the text is a non-empty string
        text = text.lower()
        text = "".join([c for c in text if c.isalpha() or c.isspace()])
    return text

df['Question'] = df['Question'].apply(preprocess_text)
df['Answer'] = df['Answer'].apply(preprocess_text)

# Drop rows with missing values in 'Question' or 'Answer'
df.dropna(subset=['Question', 'Answer'], inplace=True)


df.to_csv("data.csv")

# Load the Universal Sentence Encoder model
model_url = "https://tfhub.dev/google/universal-sentence-encoder-large/5"
embed = hub.load(model_url)

# Compute sentence embeddings for questions and answers
question_embeddings = embed(df['Question'])
answer_embeddings = embed(df['Answer'])

# Define a similarity threshold to determine if a question is relevant
similarity_threshold = 0.5

# Implement the retrieval-based chatbot function
def get_response(user_input):
    user_input = preprocess_text(user_input)
    user_embedding = embed([user_input])
    similarities = tf.reduce_sum(user_embedding * answer_embeddings, axis=1)
    most_similar_index = tf.argmax(similarities).numpy()  # Extract the integer value from the tensor
    max_similarity_score = similarities[most_similar_index].numpy()

    if max_similarity_score < similarity_threshold:
        return "I don't have an answer for that."
    else:
        return df['Answer'].iloc[most_similar_index]


user_input = input("You: ")

response = get_response(user_input)
print("Chatbot:", response)
```
